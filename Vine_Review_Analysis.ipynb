```python
import pandas as pd
```

```python
df = pd.read_csv('vine_table.csv')
```

```python
df.head()
```

```python
df_1 = df.loc[df['total_votes'] >= 20]
df_1.head()
```

```python
df_2 = df_1.loc[df_1['helpful_votes']/df_1['total_votes'] >= .5]
df_2.head()
```

```python
df_3 = df_2.loc[df_2['vine'] == 'Y']
df_3.head()
```

```python
df_4 = df_2.loc[df_2['vine'] == 'N']
df_4.head()
```

```python
total_reviews_paid = df_3['review_id'].count()
```

```python
total_reviews_unpaid = df_4['review_id'].count()
```

```python
import numpy as np
```

```python
fivestar_reviews_paid = np.count_nonzero(df_3['star_rating'] == 5)
```

```python
fivestar_reviews_unpaid = np.count_nonzero(df_4['star_rating'] == 5)
```

```python
percent_fivestar_reviews_paid = fivestar_reviews_paid/total_reviews_paid*100
```

```python
percent_fivestar_reviews_unpaid = fivestar_reviews_unpaid/total_reviews_unpaid*100
```

```python
print("The total number of paid reviews is " + str(total_reviews_paid)+".")
print("The total number of unpaid reviews is " + str(total_reviews_unpaid)+".")
print("The total number of 5 star paid reviews is " + str(fivestar_reviews_paid)+".")
print("The total number of 5 star unpaid reviews is " + str(fivestar_reviews_unpaid)+".")
print("The % of 5-star paid reviews is " + str(round(percent_fivestar_reviews_paid, 1)) +"%")
print("The % of 5-star unpaid reviews is " + str(round(percent_fivestar_reviews_unpaid,1)) +"%")
```
